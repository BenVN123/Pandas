```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv('data/school_scores.csv'))
df = df[df['Year'] == 2015]
df = df[df['State.Name'] != 'Puerto Rico']
df = df[df['State.Name'] != 'Virgin Islands']
df = df[df['State.Name'] != 'District Of Columbia']
df
```

```python
# Average math SAT score per state
df.plot.barh(x='State.Name', y='Total.Math', figsize=(30, 30))
```
