```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv('data/billionaires.csv'))
df = df[df['year'] == 2014]
df = df[df['demographics.age'] > 0]
df.head()
```

```python
# Shows the age correlation to billionaires in the world
df.plot(kind='scatter',x='demographics.age', y='wealth.worth in billions', xlabel='Wealth in Billions', alpha=0.3, color='green')
```
